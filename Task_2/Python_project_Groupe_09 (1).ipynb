visualisation de l'histogramme 

Histogramme des chambres (bedrooms):  Un histogramme est approprié pour visualiser la distribution d'une variable numérique. Dans ce cas, il montre la fréquence des différentes nombres de chambres dans les maisons du jeu de données. Cela permet de comprendre la répartition des types de logement (studios, appartements, maisons, etc.).

```python
plt.hist(data.bedrooms, bins=10, edgecolor='k')
plt.xlabel("Nombre de chambre")
plt.ylabel("Nombre de Maison")
plt.title("Distribution du nombre de chambre")
plt.show()
```

L'histogramme montre que la majorité des maisons ont 3 ou 4 chambres, avec un nombre significatif de maisons ayant 2 chambres. Il y a peu de maisons avec 5 chambres ou plus, et très peu avec 1 chambre. Cela suggère que le marché immobilier est principalement axé sur les maisons familiales de taille moyenne.

visualisation pour les nuages de points 

```python
plt.scatter(x= data.area, y = data.price )
plt.xlabel("Zone ")
plt.ylabel("Prix")
plt.title("Prix en fonction de la Zone")
plt.show()
```

Nuage de points entre la superficie et le prix:
Le nuage de points révèle une corrélation positive modérée entre la superficie et le prix des maisons. Cela signifie que les maisons plus grandes ont tendance à être plus chères. Cependant, il existe une dispersion considérable des points, ce qui indique que d'autres facteurs influencent également le prix, tels que l'emplacement, le nombre de chambres.
